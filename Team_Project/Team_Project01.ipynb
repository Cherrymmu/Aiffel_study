```python
import pandas as pd

class Human:
    dict_list = [] # self.name들을 클래스 변수에 담기 위해 빈 리스트를 만듦
    def __init__(self, dic) :
        self.name = dic # 딕셔너리 데이터를 self.name으로 불러온다.
        self.이름 = input("이름을 입력해주세요 : ") # input으로 딕셔너리에 값을 추가하는것을 보여주기 위해 추가
        self.name["이름"] = [self.이름] # str말고 list로
        Human.dict_list.append(dict(self.name)) # Class 변수에 직접 append 하기

    def show(self) :
        return len(Human.dict_list) # dict_list에 요소가 몇개 담겨져있는지 확인하는 함수

    def display_haru(self): # 각요소가 딕셔너리형태이며, 그 요소들로 하루 일정들을 출력해주는 함수
        print(f'{self.name["이름"]}의 일정:')
        for k, v in self.name.items(): # self.name에 있는 key값과 value값을 .items() 함수로 뽑아냄
            if k not in list(self.name.keys())[5:]: # 0번인덱스부터 4번인덱스까지가 일정요소이기 때문에 그뒤는 안나오게함
            # keys()를 뽑아도 딕셔너리 형태이기때문에 list()함수를 씌워 리스트로 만들어주어야 슬라이싱이 가능해집니다.
                print(f"{k}에는 {' 그리고 '.join(v)}")
            elif k in list(self.name.keys())[5]: # 5번 인덱스가 취침 시간이기때문에 elif로 이어서 조건문 실행
                print(f"{v[0]}에 잠이 듭니다.")
    def display_profile(self): # 취미부터 좋아하는계절까지 profile모양으로 출력하는 함수
        hobby_list = [] # 키워드
        value_list = [] # 키워드에 대한 활동 리스트를 받는다.

        for hobby, value in self.name.items():
            if hobby in list(self.name.keys())[6:-1] : # 이름이 추가되었기때문에, 마지막 요소를 제외하고 출력하기위해
                print(f"{hobby} : {', '.join(map(str,value))}")

    def csv_save(self) : # csv 파일로 저장하기위한 함수
        self.pd_list = [] # 각자의 csv파일을 list에 담기위해 빈리스트 생성
        for name in Human.dict_list : # Human.dict_list에 저장된 딕셔너리들을 하나씩 가져옴
            df = pd.DataFrame.from_dict(name, orient='index') # from_dict(data, orient)을 이용하여 dict형태를
                                                                # df형태로 만들어줍니다.
            df = df.transpose() # 행과 열을 뒤집어지는 함수 사용
            df["이름"] = df["이름"][0] # 이름칼럼에 input받은 value값을 넣기위해 적용
            df = df.reindex(columns=["이름", "아침", "오전", "점심", "오후", "저녁", "취침", "취미", "혈액형", "MBTI", "좋아하는 영화장르", "좋아하는 음식종류", "좋아하는 계절"])
            # 이름이 맨앞으로 나와야 csv파일에서 누구의 일정 및 소개인지 알수있기에 reindex로 새로 인덱스 부여 및 cloumns 위치 재조정
            self.pd_list.append(df) # pd_list에 append하기 위해
            display(df) # DataFrame을 print처럼 보여줄수 있다.
        self.all_df = pd.concat(self.pd_list) # pd_list에 담겨진 요소들을 한번에 column에 맞게 밑으로 합치는 함수
        self.all_df = self.all_df.reset_index() # 인덱스 새로 부여
        self.all_df.to_csv(f"전체.csv", index=False, encoding='utf-8') # csv 파일로 저장하는 함수
        display(self.all_df)


    def same_cate(self) :
        # 컬럼명들을 리스트로 배출
        # self.col_names = ["index","이름", "아침", "오전", "점심", "오후", "저녁", "취침", "취미", "혈액형", "MBTI", "좋아하는 영화장르", "좋아하는 음식종류", "좋아하는 계절"]
        self.col_names = self.all_df.columns.tolist()

        # self.col_names = ["취미", "혈액형", "MBTI", "좋아하는 영화장르", "좋아하는 음식종류", "좋아하는 계절"]
        self.col_names = self.col_names[8:]

        for i in self.col_names :
            # 공통된 요소 추출
            same_df = self.all_df[(self.all_df.duplicated([i])) & (self.all_df[i].notnull())][i]
            # df2 = df[2개이상있으면 true & null이 아니면 true]["보고싶은 col"]
            same_df2 = self.all_df[i].isin(same_df)
            # 다시 self.all_df에 있는 요소와 same_df를 비교했을때 일치하는 요소가 있는지 확인

            re_df = self.all_df[same_df2][["이름", i]]
            # 추출

            re_df_due_element = re_df[i].drop_duplicates()
            #공통된 요소들 중복 제거

            #반복문
            print()
            print(f"공통된 {i}을 가진 분들")
            for x in re_df_due_element : # re_df_due_elemnet에 담겨진 요소를 x로 뽑기
                tn_df = re_df[i].isin([x]) #참거짓 시리즈 re_df[i]에 x라는 요소가 있는지 여부 확인
                re_df_due_name = re_df[tn_df]["이름"].drop_duplicates() # 중복되는 값을 제거 후 다시 변수에 지정
                print(f"{x} : {', '.join(re_df_due_name)}") # 출력



a = Human(김민기)
b = Human(김양희)
c = Human(최강훈)
d = Human(전민규)
e = Human(이승환)
a.display_haru()
a.display_profile()
a.csv_save()
a.same_cate()
```
