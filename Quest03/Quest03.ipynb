```python
# 저희는 구글 드라이브 마운트해서 별도로 파일을 호출하는 방식을 택했습니다.
file_path = '/content/drive/MyDrive/Colab Notebooks/Avengers.txt'

with open(file_path, 'r', encoding='utf-8') as file: # r = 읽기, encoding = 글자를 불러오는 형식? 이런 개념 wxr  = 쓰기, 실행, 읽기
    content = file.read()
```

```python
import re # 특수문자 제거 정규식 제거 방법
```

```python
# lower() # 소문자로 변경하는 방법
```

```python
# 정규식 특수문자 및 괄호 제거
content_text = re.sub(r"\s*\[.*?\]", " ", content) # 대괄호제거
content_text = re.sub(r"\s*\{.*?\}", " ", content_text) # 중괄호 제거
content_text = re.sub(r'\([^)]*\)', ' ', content_text) # 소괄호 제거
content_text = re.sub(r"[^\uAC00-\uD7A30-9a-zA-Z\s]", " ", content_text) # 특수문자 제거
content_text
```

```python
# 줄바꿈이 사라지지 않아 줄바꿈 별도로 제거하기
content_text = re.sub("\n", " ", content_text)
```

```python
# 대문자 소문자로 변경하기
content_text = content_text.lower()
```

```python
content_text
```

```python
# 2gram # split 함수로 공백기준으로 단어들을 나눴습니다.
word = content_text.split()
print(word)
```

```python
word2 = list(zip(word, word[1:])) # 공백기준으로 나눈후 리스트형태이기때문에 첫번째와 두번쨰를 묶는 형태로해서 반복하여 튜플형태로 묶어주었습니다.
word2
```

```python
from collections import Counter
```

```python
word2 = Counter(word2) # zip으로 묶은 키의 갯수를 파악하기 위해 사용
word2 = sorted(word2.items(), key=lambda x: x[1], reverse=True) # value값 높은 순서대로 출력하기 위해 사용 sorted = 정렬 하는 방법, reverse 는 False = 오름차순, True = 내림차순
word2
```

```python
dict_word = dict(word2) # 다시 딕셔너리 형태로 만들기 위한 작업
```

```python
for k, v in dict_word.items(): # 키와 밸류로 나눔
    if max(dict_word.values()) == v: # 최대 밸류가 v와 동일하다면 거기에 있는 k 를 호출
        max2gram = k
# max2gram = [k for k, v in dict_word.items() if max(dict_word.values()) == v ] # 컴프리헨션 시도했다가 실패 함
# max2gram = max(dict_word.items(), key=lambda x: x[1])[0] # 람다 표현식 이건 GPT가 알려줘서 성공했음
print(max2gram, dict_word[max2gram])
Counter(dict_word)
```
